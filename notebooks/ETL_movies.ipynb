## Transformaciones

1_ Eliminar las columnas innecesarias para el desarrollo de la API, como "video", "imdb_id", "adult", "original_title", "vote_count", "poster_path" y "homepage".

2_ Rellenar con 0 los registros con valores nulos en las columnas "revenue" y "budget".

3_ Eliminar los registros que contengan valores nulos en la columna "release date".

4_ Asegurarse de que la columna "release_year" esté en formato AAAA-MM-DD y generar una nueva columna que contenga solo el año de estreno.

5_ Crear una nueva columna "return" que extraiga el retorno de inversión de cada registro, dividiendo la columna "revenue" entre la columna "budget".

6_ Desanidar los datos que se encuentran en formato de listas de diccionarios para poder unirlos nuevamente a la base de datos de manera separada.



```python
#1 Eliminamos columnas inecesarias:
df = df.drop(columns=['video', 'imdb_id','adult', 'original_title', 'poster_path','homepage'])
df.head()
```

```python
#2_ Rellenar con 0 los registros con valores nulos en las columnas "revenue" y "budget".

df["budget"] = df["budget"].fillna(0)
df["revenue"] = df["revenue"].fillna(0)
```

```python
#3_ Eliminar los registros que contengan valores nulos en la columna "release date".
df.dropna(subset=['release_date'], inplace=True)
```

```python
#4_ Asegurarse de que la columna "release_year" esté en formato AAAA-MM-DD y generar una nueva columna que contenga solo el año de estreno.
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year
```

```python
#comprobamos la nueva columna con solo el año de estreno
df['release_year']
```

```python
# 5_ Crear una nueva columna "return" que extraiga el retorno de inversión de cada registro, 
# dividiendo la columna "revenue" entre la columna "budget".

#primero pasamos a numeric la columna Budget que figura como String
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

#creamos la nueva columna 
df['return'] = df['revenue'] / df['budget']
df['return']
```

```python
# 6_ Desanidar los datos que se encuentran en formato de listas de diccionarios 
# para poder unirlos nuevamente a la base de datos de manera separada.

#creamos la funcion de convertir a diccionario el texto
def str_to_dict(text):
  try:
    return eval(text)
  except:
    return {}

#aplicamos la funcion a la columna
df['belongs_to_collection'] = df['belongs_to_collection'].apply(str_to_dict)
```

```python
#una vez como diccionario podemos aplicar pd.json_normalize para desanidar
df = pd.concat([pd.json_normalize(df['belongs_to_collection']), df.drop(['belongs_to_collection'], axis=1)], axis=1)
```
