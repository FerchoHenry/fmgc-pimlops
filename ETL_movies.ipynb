```python
import pandas as pd
import matplotlib as plt
```

```python
#carga del arhivo movies.

df = pd.read_csv("movies_dataset.csv")
df.head()
```

De un primer analisis rapido y consultando el diccionario de datos pudimos identificar algunas columnas que consideramos que no aportarian informacion util al algoritmo de recomendacion. Por ejemplo la columna homepage contiene las direcciones web de las paginas de las peliculas, la columna overview contiene una reseña del contenido de la pelicula.

```python
#Eliminamos las columnas que consideramos que no aportarian informacion util
df = df.drop(columns=['homepage', 'imdb_id','overview', 'status','overview', 'poster_path','tagline','video'])
df.head()
```

```python
df.info()
```

Notamos que de de las 45466 filas totales la columna title contiene 6 valores nulos, entonces los analizamos

```python
df[df['title'].isna()]
```

Notamos que podemos eliminar del dataset estas 6 filas

```python
# elimino los valores nan de la columna title
df = df.dropna(subset=['title'])
```

Analizamos ahora la columna adult

```python
df['adult'].value_counts()
```

Notamos 9 filas con valor TRUE las analizamos en detalle

```python
df[df['adult'] == "True"]
```

```python
df = df[df['adult'] == "False"]
```
